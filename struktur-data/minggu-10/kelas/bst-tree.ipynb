```python
class Node:
    def __init__(self, key, parent=None):
        self.key = key
        self.left = None
        self.right = None
        self.parent = parent

class BST:
    def __init__(self):
        self.root = None

    def insert(self, key):
        if self.root is None:
            self.root = Node(key)
            return
        curr = self.root
        while True:
            if key <= curr.key:
                if curr.left is None:
                    curr.left = Node(key, curr)
                    return
                curr = curr.left
            else:
                if curr.right is None:
                    curr.right = Node(key, curr)
                    return
                curr = curr.right

    def minimum(self, node):
        while node.left:
            node = node.left
        return node

    def maximum(self, node):
        while node.right:
            node = node.right
        return node

    def successor(self, node):
        if node.right:
            return self.minimum(node.right)
        curr = node
        while curr.parent and curr == curr.parent.right:
            curr = curr.parent
        return curr.parent

    def predecessor(self, node):
        if node.left:
            return self.maximum(node.left)
        curr = node
        while curr.parent and curr == curr.parent.left:
            curr = curr.parent
        return curr.parent
```
