```python
class MataKuliah:
    def __init__(self, nama):
        self.nama = nama
        self.next = None

class Mahasiswa:
    def __init__(self, nama, nim):
        self.nama = nama
        self.nim = nim
        self.matkulP = None
        self.next = None

class ListMahasiswa:
    def __init__(self):
        self.first = None

def addMahasiswa(L, M):
    if L.first is None:
        L.first = M
    else:
        temp = L.first
        while temp.next:
            temp = temp.next
        temp.next = M

def addMataKuliah(M, namaMatkul):
    newMat = MataKuliah(namaMatkul)
    if M.matkulP is None:
        M.matkulP = newMat
    else:
        temp = M.matkulP
        while temp.next:
            temp = temp.next
        temp.next = newMat

def isDuplicateCourse(head, courseName):
    temp = head
    while temp:
        if temp.nama == courseName:
            return True
        temp = temp.next
    return False

def mergeMahasiswa(list1, list2, mergedList):
    # salin mahasiswa dari list1
    temp = list1.first
    while temp:
        newM = Mahasiswa(temp.nama, temp.nim)
        # copy matkul
        mat = temp.matkulP
        while mat:
            addMataKuliah(newM, mat.nama)
            mat = mat.next
        addMahasiswa(mergedList, newM)
        temp = temp.next

    # masukkan mahasiswa dari list2
    temp = list2.first
    while temp:
        # cek apakah sudah ada di merged (NIM sama)
        cur = mergedList.first
        found = False
        while cur:
            if cur.nim == temp.nim:
                found = True
                # merge matkul
                mat = temp.matkulP
                while mat:
                    if not isDuplicateCourse(cur.matkulP, mat.nama):
                        addMataKuliah(cur, mat.nama)
                    mat = mat.next
            cur = cur.next

        if not found:  
            newM = Mahasiswa(temp.nama, temp.nim)
            mat = temp.matkulP
            while mat:
                addMataKuliah(newM, mat.nama)
                mat = mat.next
            addMahasiswa(mergedList, newM)

        temp = temp.next

def validateMergeMahasiswa(list1, list2):
    if list1.first is None and list2.first is None:
        print("Error: Kedua list kosong.")
        return False
    if list1.first is None:
        print("Error: List 1 kosong.")
        return False
    if list2.first is None:
        print("Error: List 2 kosong.")
        return False
    return True

def countTotalCourses(L):
    total = 0
    temp = L.first
    while temp:
        mat = temp.matkulP
        while mat:
            total += 1
            mat = mat.next
        temp = temp.next
    return total

def displayMergedList(L):
    temp = L.first
    while temp:
        print(f"Mahasiswa: {temp.nama} ({temp.nim})")
        print("  Mata Kuliah:")
        mat = temp.matkulP
        while mat:
            print("   -", mat.nama)
            mat = mat.next
        print()
        temp = temp.next

# MAIN
list1 = ListMahasiswa()
list2 = ListMahasiswa()
merged = ListMahasiswa()

# isi list1
m1 = Mahasiswa("Andi", "12345")
addMahasiswa(list1, m1)
addMataKuliah(m1, "Algoritma")
addMataKuliah(m1, "Struktur Data")

m2 = Mahasiswa("Budi", "67890")
addMahasiswa(list1, m2)

# isi list2
m3 = Mahasiswa("Citra", "13579")
addMahasiswa(list2, m3)
addMataKuliah(m3, "Matematika Diskrit")

# validasi + merge
if validateMergeMahasiswa(list1, list2):
    mergeMahasiswa(list1, list2, merged)

# tampilkan
displayMergedList(merged)
print("Total Mata Kuliah:", countTotalCourses(merged))
```
