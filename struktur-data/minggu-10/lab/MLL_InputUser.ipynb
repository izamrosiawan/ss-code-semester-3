```python
class MataKuliahNode:
    def __init__(self, nama_mk):
        self.nama_mk = nama_mk
        self.berikutnya = None

class DataMahasiswa:
    def __init__(self, nama_mhs, nim_mhs):
        self.nama_mhs = nama_mhs
        self.nim_mhs = nim_mhs
        self.kuliah_awal = None
        self.berikutnya = None

class KoleksiMahasiswa:
    def __init__(self):
        self.awal = None

def tambahMahasiswa(koleksi, mahasiswa):
    if koleksi.awal is None:
        koleksi.awal = mahasiswa
    else:
        penunjuk = koleksi.awal
        while penunjuk.berikutnya:
            penunjuk = penunjuk.berikutnya
        penunjuk.berikutnya = mahasiswa

def tambahMataKuliah(mahasiswa, nama_mk):
    mk_baru = MataKuliahNode(nama_mk)
    if mahasiswa.kuliah_awal is None:
        mahasiswa.kuliah_awal = mk_baru
    else:
        penunjuk = mahasiswa.kuliah_awal
        while penunjuk.berikutnya:
            penunjuk = penunjuk.berikutnya
        penunjuk.berikutnya = mk_baru

def cekDuplikatMatkul(head, nama_mk):
    penunjuk = head
    while penunjuk:
        if penunjuk.nama_mk == nama_mk:
            return True
        penunjuk = penunjuk.berikutnya
    return False

def gabungMahasiswa(kol1, kol2, hasilGabung):
    penunjuk = kol1.awal
    while penunjuk:
        mhs_baru = DataMahasiswa(penunjuk.nama_mhs, penunjuk.nim_mhs)
        mk = penunjuk.kuliah_awal
        while mk:
            tambahMataKuliah(mhs_baru, mk.nama_mk)
            mk = mk.berikutnya
        tambahMahasiswa(hasilGabung, mhs_baru)
        penunjuk = penunjuk.berikutnya

    penunjuk = kol2.awal
    while penunjuk:
        cari = hasilGabung.awal
        ditemukan = False
        while cari:
            if cari.nim_mhs == penunjuk.nim_mhs:
                ditemukan = True
                mk = penunjuk.kuliah_awal
                while mk:
                    if not cekDuplikatMatkul(cari.kuliah_awal, mk.nama_mk):
                        tambahMataKuliah(cari, mk.nama_mk)
                    mk = mk.berikutnya
            cari = cari.berikutnya

        if not ditemukan:
            mhs_baru = DataMahasiswa(penunjuk.nama_mhs, penunjuk.nim_mhs)
            mk = penunjuk.kuliah_awal
            while mk:
                tambahMataKuliah(mhs_baru, mk.nama_mk)
                mk = mk.berikutnya
            tambahMahasiswa(hasilGabung, mhs_baru)

        penunjuk = penunjuk.berikutnya

def validasiGabungMahasiswa(kol1, kol2):
    if kol1.awal is None and kol2.awal is None:
        print("Error: Kedua koleksi kosong.")
        return False
    if kol1.awal is None:
        print("Error: Koleksi 1 kosong.")
        return False
    if kol2.awal is None:
        print("Error: Koleksi 2 kosong.")
        return False
    return True

def hitungTotalMatkul(koleksi):
    total = 0
    penunjuk = koleksi.awal
    while penunjuk:
        mk = penunjuk.kuliah_awal
        while mk:
            total += 1
            mk = mk.berikutnya
        penunjuk = penunjuk.berikutnya
    return total

def tampilGabunganMahasiswa(koleksi):
    penunjuk = koleksi.awal
    while penunjuk:
        print(f"Mahasiswa: {penunjuk.nama_mhs} ({penunjuk.nim_mhs})")
        print("  Mata Kuliah:")
        mk = penunjuk.kuliah_awal
        while mk:
            print("   -", mk.nama_mk)
            mk = mk.berikutnya
        print()
        penunjuk = penunjuk.berikutnya

def inputKoleksiMahasiswa(nama_koleksi):
    koleksi = KoleksiMahasiswa()
    print(f"\nInput untuk {nama_koleksi}")

    jumlah_mhs = int(input("Jumlah mahasiswa: "))

    for _ in range(jumlah_mhs):
        nama_mhs = input("Nama mahasiswa: ")
        nim_mhs = input("NIM mahasiswa: ")

        mhs = DataMahasiswa(nama_mhs, nim_mhs)
        tambahMahasiswa(koleksi, mhs)

        jumlah_mk = int(input("Jumlah mata kuliah: "))
        for _ in range(jumlah_mk):
            nama_mk = input("  Nama matkul: ")
            tambahMataKuliah(mhs, nama_mk)

    return koleksi

# MAIN PROGRAM INTERAKTIF
kol1 = inputKoleksiMahasiswa("Koleksi 1")
kol2 = inputKoleksiMahasiswa("Koleksi 2")

hasilGabung = KoleksiMahasiswa()

if validasiGabungMahasiswa(kol1, kol2):
    gabungMahasiswa(kol1, kol2, hasilGabung)

print("\nMAHASISWA GABUNGAN")
tampilGabunganMahasiswa(hasilGabung)

print("Total Mata Kuliah:", hitungTotalMatkul(hasilGabung))
```
