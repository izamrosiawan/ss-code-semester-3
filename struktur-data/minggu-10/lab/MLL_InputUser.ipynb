```python
class MataKuliah:
    def __init__(self, nama):
        self.nama = nama
        self.next = None

class Mahasiswa:
    def __init__(self, nama, nim):
        self.nama = nama
        self.nim = nim
        self.matkulP = None
        self.next = None

class ListMahasiswa:
    def __init__(self):
        self.first = None

def addMahasiswa(L, M):
    if L.first is None:
        L.first = M
    else:
        temp = L.first
        while temp.next:
            temp = temp.next
        temp.next = M

def addMataKuliah(M, namaMatkul):
    newMat = MataKuliah(namaMatkul)
    if M.matkulP is None:
        M.matkulP = newMat
    else:
        temp = M.matkulP
        while temp.next:
            temp = temp.next
        temp.next = newMat

def isDuplicateCourse(head, courseName):
    temp = head
    while temp:
        if temp.nama == courseName:
            return True
        temp = temp.next
    return False

def mergeMahasiswa(list1, list2, mergedList):
    temp = list1.first
    while temp:
        newM = Mahasiswa(temp.nama, temp.nim)
        mat = temp.matkulP
        while mat:
            addMataKuliah(newM, mat.nama)
            mat = mat.next
        addMahasiswa(mergedList, newM)
        temp = temp.next

    temp = list2.first
    while temp:
        cur = mergedList.first
        found = False
        while cur:
            if cur.nim == temp.nim:
                found = True
                mat = temp.matkulP
                while mat:
                    if not isDuplicateCourse(cur.matkulP, mat.nama):
                        addMataKuliah(cur, mat.nama)
                    mat = mat.next
            cur = cur.next

        if not found:
            newM = Mahasiswa(temp.nama, temp.nim)
            mat = temp.matkulP
            while mat:
                addMataKuliah(newM, mat.nama)
                mat = mat.next
            addMahasiswa(mergedList, newM)

        temp = temp.next

def validateMergeMahasiswa(list1, list2):
    if list1.first is None and list2.first is None:
        print("Error: Kedua list kosong.")
        return False
    if list1.first is None:
        print("Error: List 1 kosong.")
        return False
    if list2.first is None:
        print("Error: List 2 kosong.")
        return False
    return True

def countTotalCourses(L):
    total = 0
    temp = L.first
    while temp:
        mat = temp.matkulP
        while mat:
            total += 1
            mat = mat.next
        temp = temp.next
    return total

def displayMergedList(L):
    temp = L.first
    while temp:
        print(f"Mahasiswa: {temp.nama} ({temp.nim})")
        print("  Mata Kuliah:")
        mat = temp.matkulP
        while mat:
            print("   -", mat.nama)
            mat = mat.next
        print()
        temp = temp.next

def inputListMahasiswa(namaList):
    L = ListMahasiswa()
    print(f"\nInput untuk {namaList}")

    jumlah = int(input("Jumlah mahasiswa: "))

    for _ in range(jumlah):
        nama = input("Nama mahasiswa: ")
        nim = input("NIM mahasiswa: ")

        m = Mahasiswa(nama, nim)
        addMahasiswa(L, m)

        jmlMatkul = int(input("Jumlah mata kuliah: "))
        for _ in range(jmlMatkul):
            mk = input("  Nama matkul: ")
            addMataKuliah(m, mk)

    return L

# MAIN PROGRAM INTERAKTIF
list1 = inputListMahasiswa("List 1")
list2 = inputListMahasiswa("List 2")

merged = ListMahasiswa()

if validateMergeMahasiswa(list1, list2):
    mergeMahasiswa(list1, list2, merged)

print("\nLIST MAHASISWA")
displayMergedList(merged)

print("Total Mata Kuliah:", countTotalCourses(merged))
```
