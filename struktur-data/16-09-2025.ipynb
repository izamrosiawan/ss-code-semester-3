```python
# 1. Buat sebuah list berisi nama-nama buah
buah = ["apel", "jeruk", "mangga"]
print(f"Isi list awal: {buah}")

# 2. Tambahkan buah baru di akhir
buah.append("pisang")
print(f"Setelah ditambah: {buah}")

# 3. Ambil buah di posisi kedua (indeks 1)
print(f"Buah kedua adalah: {buah[1]}")

# 4. Hapus 'jeruk' dari list
buah.remove("jeruk")
print(f"Setelah 'jeruk' dihapus: {buah}")

# 5. Cek ada berapa buah sekarang
print(f"Jumlah buah sekarang: {len(buah)}")
```
