```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class TextEditor:
    def __init__(self):
        self.head = None
        self.tail = None
        self.cursor = None

    def tambah_setelah_kursor(self, data):
        n_baru = Node(data)
        if not self.head:
            self.head = self.tail = self.cursor = n_baru
        else:
            if not self.cursor:
                n_baru.next = self.head
                self.head.prev = n_baru
                self.head = n_baru
                self.cursor = n_baru
            else:
                n_baru.prev = self.cursor
                n_baru.next = self.cursor.next
                if self.cursor.next:
                    self.cursor.next.prev = n_baru
                self.cursor.next = n_baru
                if self.cursor == self.tail:
                    self.tail = n_baru
                self.cursor = n_baru

    def hapus_di_kursor(self):
        if not self.cursor:
            return
        node_hapus = self.cursor
        if node_hapus == self.head == self.tail:
            self.head = self.tail = self.cursor = None
        elif node_hapus == self.head:
            self.head = node_hapus.next
            if self.head:
                self.head.prev = None
            self.cursor = self.head
        elif node_hapus == self.tail:
            self.tail = node_hapus.prev
            if self.tail:
                self.tail.next = None
            self.cursor = self.tail
        else:
            node_hapus.prev.next = node_hapus.next
            node_hapus.next.prev = node_hapus.prev
            self.cursor = node_hapus.prev

    def geser_kiri(self):
        if self.cursor and self.cursor.prev:
            self.cursor = self.cursor.prev

    def geser_kanan(self):
        if self.cursor and self.cursor.next:
            self.cursor = self.cursor.next

    def display(self):
        cur = self.head
        hasil = []
        while cur:
            if cur == self.cursor:
                hasil.append(f"[{cur.data}]")
            else:
                hasil.append(cur.data)
            cur = cur.next
        print('Text Editor:', ''.join(hasil))

editor = TextEditor()

while True:
    print("Menu: 1.Tambah setelah kursor 2.Hapus di kursor 3.Kiri 4.Kanan 5.Lihat 6.Keluar")
    pil = input("Pilihan: ")
    if pil == "1":
        huruf = input("Karakter: ")
        editor.tambah_setelah_kursor(huruf)
    elif pil == "2":
        editor.hapus_di_kursor()
    elif pil == "3":
        editor.geser_kiri()
    elif pil == "4":
        editor.geser_kanan()
    elif pil == "5":
        editor.display()
    elif pil == "6":
        break
```
