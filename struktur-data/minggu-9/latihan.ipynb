```python
# Single Linked List

class Node:
    def __init__(self, data):
        self.data = data      # Nilai yang disimpan di node
        self.next = None      # Penunjuk ke node berikutnya

class SingleLinkedList:
    def __init__(self):
        self.head = None      # Awal dari daftar

    # Fungsi tambah data di akhir list
    def insert(self, val):
        new_node = Node(val)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:          # Cari ekor list
                current = current.next
            current.next = new_node      # Tambahkan node baru di akhir

    # Fungsi hapus data berdasarkan nilai
    def deleteByValue(self, val):
        current = self.head
        prev = None
        while current:
            if current.data == val:
                if prev:
                    prev.next = current.next
                else:
                    self.head = current.next       # Hapus head
                print(f"Data {val} telah dihapus.")
                return
            prev = current
            current = current.next
        print("Data tidak ditemukan.")

    # Fungsi cetak isi list
    def display(self):
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        print(" -> ".join(elements))

# Contoh Pakai
sll = SingleLinkedList()
sll.insert(1)
sll.insert(2)
sll.insert(3)
sll.display()              # Output: 1 -> 2 -> 3

sll.deleteByValue(2)
sll.display()              # Output: 1 -> 3
```

```python
class Stack:
    def __init__(self):
        self.items = []
    
    def push(self, item):
        self.items.append(item)
    
    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return None
    
    def is_empty(self):
        return len(self.items) == 0
    
    def size(self):
        return len(self.items)

def membalikkan_string(s):
    stack = Stack()
    # Masukkan tiap karakter ke stack
    for char in s:
        stack.push(char)
    
    hasil = ""
    # Keluarkan dari stack (stack bersifat LIFO)
    while not stack.is_empty():
        hasil += stack.pop()
    return hasil

# Input dari pengguna
kata = input("Masukkan string yang ingin dibalik: ")
print("String terbalik:", membalikkan_string(kata))
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return None

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

def membalikkan_string(s):
    stack = Stack()
    for char in s:
        stack.push(char)
    hasil = ""
    while not stack.is_empty():
        hasil += stack.pop()
    return hasil

input_string = "contoh"
print("String terbalik:", membalikkan_string(input_string))
```
