```python
class Node:
    def __init__(self):
        self.value = None
        self.prev = None
        self.next = None

class texteditor:
    def __init__(self):
        self.head = None
        self.tail = None
        self.cursor = None
        self.size = 0

    def display(self):
        current = self.head
        while current:
            print(current.value, end=' ')
            current = current.next
        print()

    def addkarakteraftercursor(self, value):
        nodebaru = Node()
        nodebaru.value = value
        if not self.head:
            self.head = nodebaru
            self.tail = nodebaru
            self.cursor = nodebaru
        else:
            nodebaru.prev = self.cursor
            nodebaru.next = self.cursor.next
            if self.cursor.next:
                self.cursor.next.prev = nodebaru
            self.cursor.next = nodebaru
            if self.cursor == self.tail:
                self.tail = nodebaru
        self.size += 1

    def deletekarakteraftercursor(self):
        if not self.cursor or not self.cursor.next:
            return
        hapus = self.cursor.next
        self.cursor.next = hapus.next
        if hapus.next:
            hapus.next.prev = self.cursor
        if hapus == self.tail:
            self.tail = self.cursor
        self.size -= 1

    def movecursorleft(self):
        if self.cursor and self.cursor.prev:
            self.cursor = self.cursor.prev

    def movecursorright(self):
        if self.cursor and self.cursor.next:
            self.cursor = self.cursor.next



text_editor = texteditor()
user_input = input("Masukkan karakter (atau 'exit' untuk keluar): ")
while user_input.lower() != 'exit':
    text_editor.addkarakteraftercursor(user_input)
    text_editor.display()
    user_input = input("Masukkan karakter (atau 'exit' untuk keluar): ")
```

```python
class Node:
    def __init__(self, huruf):
        self.huruf = huruf
        self.prev = None
        self.next = None

class Editor:
    def __init__(self):
        self.awal = None
        self.akhir = None
        self.kursor = None

    def tampilkan(self):
        n = self.awal
        hasil = []
        while n:
            hasil.append(n.huruf)
            n = n.next
        print("Isi editor:", ''.join(hasil))
        if self.kursor:
            print(f"Kursor di: '{self.kursor.huruf}'")
        else:
            print("Kursor kosong")

    def tambah_setelah(self, huruf_baru):
        n_baru = Node(huruf_baru)
        if not self.awal:
            self.awal = self.akhir = n_baru
            self.kursor = n_baru
        else:
            if not self.kursor:
                n_baru.next = self.awal
                self.awal.prev = n_baru
                self.awal = n_baru
                self.kursor = n_baru
            else:
                n_baru.prev = self.kursor
                n_baru.next = self.kursor.next
                if self.kursor.next:
                    self.kursor.next.prev = n_baru
                self.kursor.next = n_baru
                if self.kursor == self.akhir:
                    self.akhir = n_baru
                self.kursor = n_baru

    def hapus_di_kursor(self):
        if not self.kursor:
            print("Tidak ada yang bisa dihapus.")
            return
        n_hapus = self.kursor
        if n_hapus == self.awal == self.akhir:
            self.awal = self.akhir = self.kursor = None
        elif n_hapus == self.awal:
            self.awal = n_hapus.next
            if self.awal:
                self.awal.prev = None
            self.kursor = self.awal
        elif n_hapus == self.akhir:
            self.akhir = n_hapus.prev
            if self.akhir:
                self.akhir.next = None
            self.kursor = self.akhir
        else:
            n_hapus.prev.next = n_hapus.next
            n_hapus.next.prev = n_hapus.prev
            self.kursor = n_hapus.prev

    def geser_kiri(self):
        if self.kursor and self.kursor.prev:
            self.kursor = self.kursor.prev
        else:
            print("Kursor sudah di paling kiri atau kosong.")

    def geser_kanan(self):
        if self.kursor and self.kursor.next:
            self.kursor = self.kursor.next
        else:
            print("Kursor sudah di paling kanan atau kosong.")

editor_teks = Editor()
```
