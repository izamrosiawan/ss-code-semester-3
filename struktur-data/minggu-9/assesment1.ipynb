```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
    
class texteditor:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def append(self, value):
        new_node = Node(value)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.size += 1

    def display(self):
        current = self.head
        while current:
            print(current.value, end=' ')
            current = current.next
        print()
    
    
```
