SingleLinkedList

```python
# Node punya info sama penunjuk ke node berikutnya
class Node:
    def __init__(self, data):
        self.info = data
        self.next = None

# List ini wadah buat nyimpen node-node tadi
class List:
    def __init__(self):
        self.first = None

    # Biar listnya kosong
    def createList(self):
        self.first = None

    # Bikin node baru pake data x
    def createElement(self, x):
        return Node(x)

    # Masukin node P ke paling depan
    def insertFirst(self, P):
        if P:
            P.next = self.first
            self.first = P

    # Masukin node P ke paling belakang
    def insertLast(self, P):
        if P:
            if not self.first:
                self.first = P
            else:
                last = self.first
                while last.next:
                    last = last.next
                last.next = P

    # Nampilin semua isi list dari depan ke belakang
    def printList(self):
        if not self.first:
            print("List kosong")
        else:
            current = self.first
            while current:
                print(current.info, end=" ")
                current = current.next
            print()

    # Hapus node paling depan
    def deleteFirst(self):
        if self.first:
            P = self.first
            self.first = P.next
            return P
        return None

    # Hapus node paling belakang
    def deleteLast(self):
        if not self.first:
            return None
        if not self.first.next:
            P = self.first
            self.first = None
            return P
        current = self.first
        while current.next.next:
            current = current.next
        P = current.next
        current.next = None
        return P

    # Cari node yang info-nya sama dengan x
    def searchInfo(self, x):
        current = self.first
        while current and current.info != x:
            current = current.next
        return current

    # Balik urutan listnya, jadi yang tadinya paling belakang jadi paling depan
    def reverseList(self):
        last_node = None
        current = self.first
        while current:
            next_node = current.next
            current.next = last_node
            last_node = current
            current = next_node
        self.first = last_node

    # Nampilin isi list dari belakang ke depan
    def printReverse(self):
        def _printReverse(node):
            if node:
                _printReverse(node.next)
                print(node.info, end=" ")
        _printReverse(self.first)
        print()

    # Masukin data ke list lewat input
    def inputList(self):
        n = int(input("Masukkan jumlah elemen: "))
        for i in range(n):
            x = int(input("Masukkan elemen: "))
            self.insertLast(self.createElement(x))

# Bikin list kosong, isi data, print, dibalik, print lagi
main = List()
main.createList()
main.inputList()
main.printList()
main.reverseList()
main.printList()
```

DoubleLinkedList

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # Masukin elemen ke belakang
    def insertLast(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        new_node.prev = last

    # Hapus elemen berdasarkan nilai
    def deleteByValue(self, value):
        if not self.head:
            print("List kosong")
            return
        current = self.head
        found = False
        while current:
            if current.data == value:
                found = True
                # Kalau node di tengah/tengah akhir
                if current.prev:
                    current.prev.next = current.next
                else:
                    # Node di depan
                    self.head = current.next
                if current.next:
                    current.next.prev = current.prev
                print(f"Elemen {value} berhasil dihapus")
                break
            current = current.next
        if not found:
            print(f"Elemen {value} tidak ditemukan")

    # Nampilin isi list
    def printList(self):
        current = self.head
        if not current:
            print("List kosong")
            return
        while current:
            print(current.data, end="")
            if current.next:
                print(" – ", end="")
            current = current.next
        print()

# Main program
dll = DoublyLinkedList()
inp = input("Masukkan elemen (pisahkan dengan koma, misal 5,10,15): ")
numbers = [int(x.strip()) for x in inp.split(",")]
for num in numbers:
    dll.insertLast(num)

dll.printList()
hapus = int(input("Masukkan nilai yang ingin dihapus: "))
dll.deleteByValue(hapus)
dll.printList()
```
