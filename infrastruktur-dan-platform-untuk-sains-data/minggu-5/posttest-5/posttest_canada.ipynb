```python
# Contoh: Visualisasi
import matplotlib.pyplot as plt

# Misal: Plot imigran per tahun
if 'year' in data.columns:
    data.groupby('year').sum().plot(y='jumlah_imigran', kind='line')
    plt.title('Jumlah Imigran ke Kanada per Tahun')
    plt.ylabel('Jumlah Imigran')
    plt.show()
else:
    print('Kolom "year" tidak ditemukan, sesuaikan dengan struktur datamu.')
```
