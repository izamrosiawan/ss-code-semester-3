```python
query = """
SELECT
    sr.sales_outlet_id,
    g.generation,
    COUNT(*) AS jumlah_pesanan
FROM
    'sales reciepts' sr
    JOIN customer c ON sr.customer_id = c.customer_id
    JOIN generations g ON c.birth_year = g.birth_year
WHERE
    sr.sales_outlet_id IN (
        SELECT sr2.sales_outlet_id
        FROM 'sales reciepts' sr2
        GROUP BY sr2.sales_outlet_id
        ORDER BY COUNT(*) DESC
        LIMIT 3
    )
GROUP BY
    sr.sales_outlet_id, g.generation
ORDER BY
    sr.sales_outlet_id, jumlah_pesanan DESC
"""
df = pd.read_sql_query(query, conn)
print(df)
```

```python
#Produk Paling Sering Dipesan Setiap Generasi
query = """
SELECT 
    g.generation, 
    p.product, 
    COUNT(*) AS jumlah_dipesan
FROM 
    'sales reciepts' sr
    JOIN customer c ON sr.customer_id = c.customer_id
    JOIN generations g ON c.birth_year = g.birth_year
    JOIN product p ON sr.product_id = p.product_id
GROUP BY 
    g.generation, p.product
ORDER BY 
    g.generation, jumlah_dipesan DESC
"""
df = pd.read_sql_query(query, conn)
print(df)
```

```python
#Total Penjualan per Kategori Produk
query = """
SELECT 
    p.product_category, 
    SUM(sr.quantity) AS total_terjual
FROM 
    'sales reciepts' sr
    JOIN product p ON sr.product_id = p.product_id
GROUP BY 
    p.product_category
ORDER BY 
    total_terjual DESC
"""
df = pd.read_sql_query(query, conn)
print(df)
```

```python
#Jumlah Pelanggan per Generasi
query = """
SELECT 
    g.generation, 
    COUNT(DISTINCT c.customer_id) AS jumlah_pelanggan
FROM 
    customer c
    JOIN generations g ON c.birth_year = g.birth_year
GROUP BY 
    g.generation
ORDER BY 
    jumlah_pelanggan DESC
"""
df = pd.read_sql_query(query, conn)
print(df)
```

```python
#Outlet dengan Penjualan Terendah
query = """
SELECT 
    sr.sales_outlet_id, 
    COUNT(*) AS jumlah_transaksi
FROM 
    'sales reciepts' sr
GROUP BY 
    sr.sales_outlet_id
ORDER BY 
    jumlah_transaksi ASC
LIMIT 1
"""
df = pd.read_sql_query(query, conn)
print(df)
```

```python
#Produk yang Tidak Pernah Dipesan
query = """
SELECT 
    p.product_id, 
    p.product
FROM 
    product p
LEFT JOIN 
    'sales reciepts' sr ON p.product_id = sr.product_id
WHERE 
    sr.product_id IS NULL
"""
df = pd.read_sql_query(query, conn)
print(df)
```

```python
#Perbandingan Penjualan Produk Retail vs Warehouse
query = """
SELECT 
    so.sales_outlet_type, 
    COUNT(sr.transaction_id) AS jumlah_transaksi
FROM 
    'sales reciepts' sr
    JOIN 'sales outlet' so ON sr.sales_outlet_id = so.sales_outlet_id
GROUP BY 
    so.sales_outlet_type
"""
df = pd.read_sql_query(query, conn)
print(df)
```

```python
#Jumlah Transaksi Per Hari
query = """
SELECT 
    transaction_date, 
    COUNT(*) AS jumlah_transaksi
FROM 
    'sales reciepts'
GROUP BY 
    transaction_date
ORDER BY 
    transaction_date
"""
df = pd.read_sql_query(query, conn)
print(df)
```

```python
#Rata-rata Umur Pelanggan per Generasi
query = """
SELECT 
    g.generation, 
    AVG(2019 - c.birth_year) AS rata2_umur
FROM 
    customer c
    JOIN generations g ON c.birth_year = g.birth_year
GROUP BY 
    g.generation
"""
df = pd.read_sql_query(query, conn)
print(df)
```

```python
#5 Pelanggan dengan Transaksi Terbanyak
query = '''
SELECT
    c.customer_id,
    c."customer_first-name" AS customer_first_name,
    COUNT(sr.transaction_id) AS jumlah_transaksi
FROM
    "sales reciepts" sr
    JOIN customer c ON sr.customer_id = c.customer_id
GROUP BY
    c.customer_id
ORDER BY
    jumlah_transaksi DESC
LIMIT 5
'''
df = pd.read_sql_query(query, conn)
print(df)
```
