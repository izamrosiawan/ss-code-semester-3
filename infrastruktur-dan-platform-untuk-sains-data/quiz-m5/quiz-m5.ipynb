generasi mana yang memiliki loyalitas paling tinggi

```python

# Generasi dengan loyalitas tertinggi berdasarkan lama kartu loyalitas pelanggan
query = """
SELECT 
    g.generation,
    COUNT(c.customer_id) AS jumlah_pelanggan,
    AVG(JULIANDAY('2019-12-31') - JULIANDAY(c.customer_since)) / 365.25 AS rata_tahun_loyalitas
FROM customer c
JOIN generations g ON c.birth_year = g.birth_year
GROUP BY g.generation
ORDER BY rata_tahun_loyalitas DESC
"""

df_loyalty = pd.read_sql_query(query, conn)
print(df_loyalty)
```
